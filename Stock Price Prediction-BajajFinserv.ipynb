```python
#split the datat into independent and dependent variable
x=df.drop(['Close'],axis=1)
y=df['Close']
```

```python
#build the linear regression modelfor given datatset
from sklearn.linear_model import LinearRegression
```

```python
reg= LinearRegression()
```

```python
df.tail()
```

```python
#split the data into training and testing 

from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=1)
```

```python
from sklearn.preprocessing import StandardScaler
#create the object of StandardScaler class
ss=StandardScaler()
X_train=ss.fit_transform(X_train) #fit is used to train the model
X_test=ss.transform(X_test)
```

```python
model=reg.fit(X_train,y_train)
```

```python
pred=model.predict(X_test)
```

```python
print(pred)
```

```python
#To check the mean squred error and R2 score for the accuracy of model
from sklearn.metrics import mean_squared_error,r2_score
```

```python
#find the mean squared error
print("Mean Squared error : ",mean_squared_error(y_test,pred))
#find the r2_score 
print("R2 Score : ",r2_score(y_test,pred))
print("Training score : ",model.score(X_train,y_train))
print("Testing score : ",model.score(X_test,y_test))
```
